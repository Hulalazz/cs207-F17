```python
from pathlib import Path
import importlib.util
import threading
import pyautogui
import time
# parent directory of student repos
d = '/Users/charlesliu/Documents/HarvardTeaching/CS207/studentrepos'
```

```python
# aggregate student repos in directory
p = Path(d)
studentrepos = []
for x in p.iterdir():
    repo = str(x).replace(d + '/', '').lower()
    if 'cs207' in repo:
        studentrepos.append((x, repo))
```

```python
# tests to run given module, return score
s = 'ABCDEFGHI'
def stoi(arr):
    return list(map(lambda x: s.index(x), arr))
input_list = stoi(['F', 'B', 'A', 'D', 'C', 'E', 'G', 'I', 'H'])
inorder = stoi(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'])
postorder = stoi(['A', 'C', 'E', 'D', 'B', 'H', 'I', 'G', 'F'])
preorder = stoi(['F', 'B', 'A', 'D', 'C', 'E', 'G', 'I', 'H'])
def testSuite (m):
    score = 1
    incorrect = []
    try:
        try:
            b = m.BinaryTree()
        except:
            try:
                b = m.BinarySearchTree()
            except:
                b = m.BST()
        for val in input_list:
            b.insert(val)
    except Exception as e:
        return '1 (' + str(e) + ')'
    
    try:
        traversal = m.DFSTraversal(b, m.DFSTraversalTypes.INORDER)
        for traversalType in m.DFSTraversalTypes:
            idx = 0
            correct = True
            yourVal = ''
            correctVal = ''
            traversal.changeTraversalType(traversalType)
            try:
                for val in traversal:
                    if traversalType.name == 'INORDER':
                        if inorder[idx] != val:
                            correct = False
                            correctVal = inorder[idx] 
                            yourVal = val
                            break
                    elif traversalType.name == 'POSTORDER':
                        if postorder[idx] != val:
                            correct = False
                            correctVal = postorder[idx]
                            yourVal = val
                            break
                    else:
                        if preorder[idx] != val:
                            correct = False
                            correctVal = preorder[idx]
                            yourVal = val
                            break
                    idx += 1
                if correct:
                    score += 1
                else:
                    incorrect.append(traversalType.name + ': ' + str(idx) + '=' + str(yourVal) + ',' + str(correctVal))
            except Exception as e:
                incorrect.append(traversalType.name + ': ' + str(e))
    except Exception as e:
        return '1 (' + str(e) + ')'
    
    return str(score) + (' (' + str(incorrect) + ')' if score < 4 else '')
```

```python
# run script
grades = {}
for path, repo in studentrepos:
    hw8 = path / 'homeworks' / 'HW8' / 'TreeTraversal.py'
    try:
        spec = importlib.util.spec_from_file_location("TreeTraversal", str(hw8.resolve()))
        b = importlib.util.module_from_spec(spec)
        spec.loader.exec_module(b)
        grades[repo] = testSuite(b)
    except Exception as e:
        grades[repo] = '1 (' + str(e) + ')'
grades
```
